```python
#largest element in array O(n)

def find_largest(arr):
    # 1. Initialize max_val with the first element
    max_val = arr[0] 
    
    # 2. Loop through the rest of the array
    for num in arr:
        # 3. If current number is larger than our current max, update it
        if num > max_val:
            max_val = num
            
    return max_val

numbers = [10, 24, 76, 23, 12]
print(find_largest(numbers))  # Output: 76

#method 2 O(nlogn)
arr = [10, 24, 76, 23, 12]
arr.sort()
largest = arr[-1]
print(largest)

```

```python
#Second largest element  O(n)
def find_second_extremes(arr):
    # Handle edge case: list must have at least 2 unique elements
    if len(arr) < 2:
        return "Array needs at least two unique elements"

    # Initialize with extreme values
    largest = second_largest = float('-inf')
    smallest = second_smallest = float('inf')

    for num in arr:
        # --- Update Largest & Second Largest ---
        if num > largest:
            second_largest = largest
            largest = num
        elif num > second_largest and num != largest:
            second_largest = num

        # --- Update Smallest & Second Smallest ---
        if num < smallest:
            second_smallest = smallest
            smallest = num
        elif num < second_smallest and num != smallest:
            second_smallest = num

    return second_largest, second_smallest

# Example Usage:
numbers = [12, 35, 1, 10, 34, 35, 1]
sec_max, sec_min = find_second_extremes(numbers)

print(f"Second Largest: {sec_max}")  # Output: 34
print(f"Second Smallest: {sec_min}") # Output: 10
```

```python
#check array is sorted O(n)

def is_sorted(arr):
    # An empty array or single element array is always sorted
    if len(arr) <= 1:
        return True

    # Check if sorted in ascending order
    is_ascending = True
    for i in range(len(arr) - 1):
        if arr[i] > arr[i + 1]:
            is_ascending = False
            break  # Exit loop early if ascending order fails

    # Check if sorted in descending order
    is_descending = True
    for i in range(len(arr) - 1):
        if arr[i] < arr[i + 1]:
            is_descending = False
            break  # Exit loop early if descending order fails

    return is_ascending or is_descending

# Example Usage:
print(is_sorted([1, 5, 8, 12]))    # Output: True (Ascending)
print(is_sorted([9, 7, 4, 1]))     # Output: True (Descending)
print(is_sorted([2, 8, 3, 10]))    # Output: False (Unsorted)
print(is_sorted([5, 5, 5]))        # Output: True (Equal elements)
```

```python
#To remove duplicates from a sorted array in-place O(n)

def remove_duplicates(arr, n):
    # Handle edge case for empty array
    if n == 0:
        return 0
        
    i = 0
    # Loop from index 1 up to n-1 (corresponds to for(int j = 1; j < n; j++))
    for j in range(1, n):
        if arr[i] != arr[j]:
            arr[i + 1] = arr[j]
            i += 1
            
    # Return the count of unique elements
    return i + 1

# Example Usage:
nums = [1, 2, 2, 2, 3, 4, 4]
size = len(nums)

unique_count = remove_duplicates(nums, size)

print(f"Number of unique elements: {unique_count}") # Output: 4
print(f"Modified array (first 4 items): {nums[:unique_count]}") # Output: [1, 2, 3, 4]
```

```python
#Left rotate array by one place O(n)
def rotate(a):
  temp = a[0]

  for i in range(1, len(a)):
      a[i-1]=a[i]

  a[len(a)-1] = temp

  return a

a = [1,2,3,4,5]
print(rotate(a))


#Method 2 O(n)
def rotate_clean(a):
    #a[1:] gets everything from index 1 to the end: [2, 3, 4, 5].
    #a[:1] gets just the first element inside a list: [1].
    return a[1:] + a[:1]
    #a[i:] + a[:i] rotate by i places

a = [1, 2, 3, 4, 5]
print(rotate_clean(a))  # Output: [2, 3, 4, 5, 1]

```

```python
#Left Rotate array by d places O(n)

def reverse(arr, start, end):
    # Helper function to reverse a segment of the array in-place
    while start < end:
        arr[start], arr[end] = arr[end], arr[start]
        start += 1
        end -= 1

def rotate_by_d(arr, d):
    n = len(arr)
    if n == 0:
        return arr
        
    # Handle cases where d is greater than the array length
    d = d % n  # 3%7 = 3
    #If an array has 5 elements, rotating it by 5 places results in the exact same array.
    #Rotating it by 6 places is identical to rotating it by 1 place.

    # Step 1: Reverse the first d elements
    reverse(arr, 0, d - 1)
    
    # Step 2: Reverse the remaining n-d elements
    reverse(arr, d, n - 1)
    
    # Step 3: Reverse the entire array
    reverse(arr, 0, n - 1)
    
    return arr

# Example Usage:
nums = [1, 2, 3, 4, 5, 6, 7]
print(rotate_by_d(nums, 3))  # Output: [4, 5, 6, 7, 1, 2, 3]

```

```python
# Right rotate array by  d element

def reverse(arr, start, end):
    while start < end:
        arr[start], arr[end] = arr[end], arr[start]
        start += 1
        end -= 1

def right_rotate_by_d(arr, d):
    n = len(arr)
    if n == 0:
        return arr
        
    d = d % n 

    # Step 1: Reverse the last 'd' elements
    reverse(arr, n - d, n - 1)
    
    # Step 2: Reverse the first 'n - d' elements
    reverse(arr, 0, n - d - 1)
    
    # Step 3: Reverse the entire array
    reverse(arr, 0, n - 1)
    
    return arr

# Example Usage:
nums = [1, 2, 3, 4, 5, 6, 7]
print(right_rotate_by_d(nums, 3))  # Output: [5, 6, 7, 1, 2, 3, 4]
```

```python
# Move all zero to the end of array O(n)

def move_zeroes(arr):
    # Pointer to track where the next non-zero element should go
    non_zero_pos = 0
    
    # Iterate through the array
    for i in range(len(arr)):
        if arr[i] != 0: # check only for non zero no
            # Swap elements
            arr[non_zero_pos], arr[i] = arr[i], arr[non_zero_pos]
            non_zero_pos += 1
            
    return arr

# Example Usage:
nums = [1, 0, 2, 3, 0, 4, 0, 5]
print(move_zeroes(nums))  # Output: [1, 2, 3, 4, 5, 0, 0, 0]
```

```python
# Linear Search O(n)

def linear_search(arr, target):
    # Loop through every element and its index
    for i in range(len(arr)):
        if arr[i] == target:
            return i  # Return the index immediately if found
            
    return -1  # Return -1 if the target does not exist

# Example Usage:
nums = [10, 23, 45, 70, 11, 15]
target_val = 70

result = linear_search(nums, target_val)
print(f"Element found at index: {result}")  # Output: 3
```

```python
#union of two sorted array O(n1 + n2)

def find_union(a, b):
    n1, n2 = len(a), len(b)
    i, j = 0, 0  # i for array a, j for array b
    union = []   # List to store the result

    # Loop through both arrays simultaneously
    while i < n1 and j < n2:
        # Case 1: Elements are equal
        if a[i] == b[j]:
            if len(union) == 0 or union[-1] != a[i]:
                union.append(a[i])
                #union[-1] --> latest last element union array
                #a[i] --> next new element to add a array
            i += 1
            j += 1
            
        # Case 2: Element in 'a' is smaller
        elif a[i] < b[j]:
            if len(union) == 0 or union[-1] != a[i]:
                union.append(a[i])
            i += 1
            
        # Case 3: Element in 'b' is smaller
        else:
            if len(union) == 0 or union[-1] != b[j]:
                union.append(b[j])
            j += 1

    # Add remaining elements from array a (if any)
    while i < n1:
        if union[-1] != a[i]:
            union.append(a[i])
        i += 1

    # Add remaining elements from array b (if any)
    while j < n2:
        if union[-1] != b[j]:
            union.append(b[j])
        j += 1

    return union

# Example Usage:
arr1 = [1, 2, 3, 4, 5, 6]
arr2 = [2, 3, 5, 7]
print(find_union(arr1, arr2))  # Output: [1, 2, 3, 4, 5, 6, 7]


#Method 2 O(nlogn)

def find_union_quick(a, b):
    return sorted(list(set(a) | set(b)))

arr1 = [1, 2, 3, 4, 5, 5, 6]
arr2 = [2, 3, 5, 7]
print(find_union_quick(arr1, arr2))  # Output: [1, 2, 3, 4, 5, 6, 7]
```

```python
#Intersection of two sorted arrays O(n1 + n2)

def find_intersection(a, b):
    n1, n2 = len(a), len(b)
    i, j = 0, 0        # i for array a, j for array b
    intersection = []  # List to store the shared elements

    # Loop through both arrays simultaneously
    while i < n1 and j < n2:
        # Case 1: Elements match! We found a common item
        if a[i] == b[j]:
            # Duplicate check: Only add if it's the first item or different from the last added item
            if len(intersection) == 0 or intersection[-1] != a[i]:
                intersection.append(a[i])
            i += 1
            j += 1
            
        # Case 2: Element in 'a' is smaller, so it can't match b[j]
        elif a[i] < b[j]:
            i += 1
            
        # Case 3: Element in 'b' is smaller, so it can't match a[i]
        else:
            j += 1

    return intersection

# Example Usage:
arr1 = [1, 2, 2, 3, 4]
arr2 = [2, 2, 4, 5, 6]
print(find_intersection(arr1, arr2))  # Output: [2, 4]




#Method 2 O(n1 + n2 + klogk)

def find_intersection_quick(a, b):
    return sorted(list(set(a) & set(b)))

arr1 = [1, 2, 2, 3, 4]
arr2 = [2, 2, 4, 5, 6]
print(find_intersection_quick(arr1, arr2))  # Output: [2,4]
```

```python
#Find missing number in an array O(n)

def find_missing_number(arr, n):
    # 1. Calculate sum of numbers from 1 to n using the mathematical formula
    expected_sum = (n * (n + 1)) // 2
    
    # 2. Calculate the actual sum of elements present in the array
    actual_sum = sum(arr)
    
    # 3. The difference is the missing number
    return expected_sum - actual_sum

# Example Usage:
# An array of size 4 containing numbers from 1 to 5, where 4 is missing
nums = [1, 2, 5, 3] 
total_range_elements = 5 # N value

print(find_missing_number(nums, total_range_elements)) # Output: 4


#Method 2 XOR  A ^ A = 0 O(n)
def missing_number(a, N):
    xor1 = 0
    xor2 = 0
    
    # n is the size of the given array (N - 1)
    n = N - 1
    
    # Single loop to calculate both XORs simultaneously
    for i in range(n):
        xor2 = xor2 ^ a[i]      # XOR of array elements
        xor1 = xor1 ^ (i + 1)  # XOR of numbers from 1 to N-1
        
    # XOR the last number N which was left out by the loop
    xor1 = xor1 ^ N
    
    # The final answer is the XOR combination of both
    return xor1 ^ xor2

# Example Usage:
# A range of 1 to 5 (N=5), array size is 4, missing number is 4
nums = [1, 2, 5, 3] 
total_elements = 5

print(missing_number(nums, total_elements))  # Output: 4

```

```python
#Maximum consecutive one in array O(n) i.e find continous one

def find_max_consecutive_ones(arr):
    max_ones = 0      # Stores the overall global maximum
    current_ones = 0  # Stores the current consecutive streak

    for num in arr:
        if num == 1:
            current_ones += 1  # Extend the current streak
            # Update max_ones if the current streak beats the record
            if current_ones > max_ones:
                max_ones = current_ones
        else:
            current_ones = 0   # Streak broken! Reset counter to 0

    return max_ones

# Example Usage:
nums = [1, 1, 0, 1, 1, 1, 0, 1]
print(find_max_consecutive_ones(nums))  # Output: 3
```

```python
#Find the number that appears once, and other numbers twice. O(n) 

def get_single_element(arr):
    xor_sum = 0
    
    # XOR all elements in the array
    for num in arr:
        xor_sum = xor_sum ^ num
        
    return xor_sum

# Example Usage:
nums = [4, 1, 2, 1, 2]
print(get_single_element(nums))  # Output: 4
```
